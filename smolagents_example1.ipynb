```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
from IPython.display import display

# Load a pre-trained sentence embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Define a reference sentence and progressively modified sentences
sentences = [
    "A cat is sitting on the windowsill.",  # Reference sentence
    "A feline is resting on the windowsill.",  # Synonym substitution (minimal change)
    "A cat is sitting by the window.",  # Slight rewording
    "A dog is sitting on the windowsill.",  # One key word changed (species swap)
    "A cat is running through the garden.",  # Action & setting changed
    "The book lies open on the table.",  # Completely different topic
]

# Compute embeddings
embeddings = [model.encode(sentence).reshape(1, -1) for sentence in sentences]

# Compute cosine similarity to the reference sentence
reference_embedding = embeddings[0]
similarities = [cosine_similarity(reference_embedding, emb)[0][0] for emb in embeddings]

# Create a dataframe for visualization
df = pd.DataFrame({"Sentence": sentences, "Cosine Similarity": similarities})
display(df)
```
